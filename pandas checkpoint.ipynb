```python
import pandas as pd
import numpy as np
#question 1
```

```python
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5,np.nan   , 9, 20, 14.5, np.nan  , 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df=pd.DataFrame(exam_data, index=labels)
df.head(3)
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
x=df[["name","score"]]
x
```

```python
dict2={"name" : ["Suresh"], "score": [15.5], "attempts": [1], "qualify": ["yes"]}
df1=pd.DataFrame(dict2)
df1
```

```python
new_df=pd.concat([df,df1])
new_df
```

```python
new_df.drop("attempts",axis=1,inplace=True)
new_df
```

```python
Success= []
for row in new_df['score']:
    if row > 10 :
        Success.append(1)
    else:
        Success.append(0)

new_df['Success'] =Success
new_df.insert(4,"Success1",Success)
new_df
```

```python
new_df.to_csv("question1.csv",index=False)
```
